```python
# Instale as bibliotecas necessárias
!pip install SpeechRecognition
!pip install gtts
!pip install pydub

import speech_recognition as sr
from gtts import gTTS
from IPython.display import Audio, display
import webbrowser

# Função para reconhecer o comando de texto
def reconhecer_comando():
    comando = input("Digite o comando: ")
    return comando.lower()

# Função para executar o comando
def executar_comando(comando):
    if "open wikipedia" in comando:
        webbrowser.open("https://www.wikipedia.org")
        resposta = "Abrindo Wikipedia..."
    else:
        resposta = "Comando não reconhecido."

    # Converter texto em fala
    tts = gTTS(resposta, lang='pt')
    tts.save("resposta.mp3")
    display(Audio("resposta.mp3", autoplay=True))

# Loop principal
while True:
    comando = reconhecer_comando()
    if comando:
        executar_comando(comando)
```
